```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import mutual_info_classif

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
target_map = {'Yes':1, 'No':0}
Y = dataset['Attrition'].apply(lambda x: target_map[x])
X = pd.get_dummies(dataset.drop(['Attrition'], axis=1), drop_first=True)  # Over18 gets dropped.
```

```python
X.head()
```

# Feature Selection

```python
# Removing Columns having 0 Variance

var_thres = VarianceThreshold(threshold=0)
var_thres.fit(X)
constant_columns = [column for column in X.columns
                   if column not in X.columns[var_thres.get_support()]]
X = X.drop(constant_columns, axis=1)
```

```python
X.head()
```

```python
plt.figure(figsize=(15, 15))
cor = X.corr()
sns.heatmap(cor, annot=False, cmap=plt.cm.CMRmap_r)
plt.show()
```

Mutual Information

```python
# Splitting data to avoid overfitting
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, stratify=Y, random_state=0)
```

```python
mutual_info = mutual_info_classif(x_train, y_train)
mutual_info = pd.Series(mutual_info)
mutual_info.index = x_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```
